```python
import pandas as pd
import numpy as np
data=pd.read_csv("数据集\\house.csv")
data.head()
```

```python
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='HarmonyOS Sans SC'
fig=plt.figure(figsize=(16,8))
fig1=plt.subplot(121)
plt.scatter(data.loc[:,'面积'],data.loc[:,'价格'])
plt.xlabel('面积')
plt.ylabel('价格')
plt.title('面积-价格')
fig2=plt.subplot(122)
plt.scatter(data.loc[:,'房间数量'],data.loc[:,'价格'])
plt.xlabel('房间数量')
plt.ylabel('价格')
plt.title('房间数量-价格')
plt.show()
```
