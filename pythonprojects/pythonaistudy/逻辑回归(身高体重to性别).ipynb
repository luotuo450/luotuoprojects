```python
import pandas as pd
data=pd.read_csv("数据集\\weight and height to gender.csv")
data.head()
```

```python
x=data.drop(['gender'],axis=1).values
y=data.loc[:,'gender'].values
x1=data.loc[:,'height']
x2=data.loc[:,'weight']
```

```python
mask=data.loc[:,"gender"]==1
```

```python
from matplotlib import pyplot as plt
fig=plt.figure()
man=plt.scatter(x1[mask],x2[mask])
woman=plt.scatter(x1[~mask],x2[~mask],marker='^')
plt.legend((man,woman),('man','woman'))
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height-weight')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x,y)
```

```python
y_pred=model.predict([[151,35]])
print(y_pred)
```
